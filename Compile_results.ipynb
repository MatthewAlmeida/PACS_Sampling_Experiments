```python
SDS_DIR = Path(os.getcwd())/"results/sds"
NOSDS_DIR = Path(os.getcwd())/"results/nosds"

SDS_DIR, NOSDS_DIR
```

```python
def get_accuracies_from_CM_output(results_dir_path, n_labels=7):
    accuracies = []
    test_set_size = 0

    for filename in os.scandir(results_dir_path):
        CM = torch.load(Path(filename)).reshape((7,7))
    
        if test_set_size == 0:
            test_set_size = torch.sum(CM)
    
        acc = torch.trace(CM)/ test_set_size
        acc=acc.item()
        accuracies.append(acc)
    
    return accuracies
```

```python
sds_accuracies = get_accuracies_from_CM_output(SDS_DIR)
nosds_accuracies = get_accuracies_from_CM_output(NOSDS_DIR)
```

```python
plt.hist(sds_accuracies)
plt.show()
```
